```python
import pandas as pd
train = pd.read_csv('datasets/breast-cancer-train.csv', usecols=['Clump Thickness', 'Cell Size', 'Type'])
X_train = train[['Clump Thickness', 'Cell Size']]
y_train = train[['Type']]

test = pd.read_csv('datasets/breast-cancer-test.csv', usecols=['Clump Thickness', 'Cell Size', 'Type'])
X_test = test[['Clump Thickness', 'Cell Size']]
y_test = test[['Type']]

test_negative = test.loc[test['Type'] == 0][['Clump Thickness', 'Cell Size', 'Type']]
test_positive = test.loc[test['Type'] == 1][['Clump Thickness', 'Cell Size', 'Type']]
```

```python
plt.scatter(test_negative['Clump Thickness'],test_negative['Cell Size'], marker = 'o', s=200, c='green', label='benign')
plt.scatter(test_positive['Clump Thickness'],test_positive['Cell Size'], marker = 'x', s=200, c='red', label='malignant')
plt.xlabel('Clump Thickness')
plt.ylabel('Cell Size')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300 

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr = LogisticRegression()

lr.fit(X_train, y_train)

y_predict = lr.predict(X_test)

accuracy_score(y_test, y_predict)
```

```python
lx = np.arange(0, 12)
ly = (-lr.intercept_ - lx * lr.coef_[0,0]) / lr.coef_[0,1]
```

```python
plt.plot(lx, ly, c='black')
plt.scatter(test_negative['Clump Thickness'],test_negative['Cell Size'], marker = 'o', s=200, c='green', label='benign')
plt.scatter(test_positive['Clump Thickness'],test_positive['Cell Size'], marker = 'x', s=200, c='red', label='malignant')
plt.xlabel('Clump Thickness')
plt.ylabel('Cell Size')
plt.legend()
plt.show()
```

```python
lr = LogisticRegression()

lr.fit(X_train[:10], y_train[:10])

y_predict = lr.predict(X_test)

accuracy_score(y_test, y_predict)
```
